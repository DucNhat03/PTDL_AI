```python
# Import các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Tải bộ dữ liệu Boston Housing
from sklearn.datasets import fetch_openml
boston = fetch_openml(name='boston', version=1, as_frame=True)
df = boston.frame

# Chọn các đặc trưng 'AGE' và 'LSTAT' và biến mục tiêu 'MEDV'
X = df[['AGE', 'LSTAT']]
y = df['MEDV']

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra với tỷ lệ 80:20 và random_state=10
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Áp dụng log transformation lên các đặc trưng trong tập huấn luyện
X_train_log = np.log(X_train)

# Áp dụng log transformation lên các đặc trưng trong tập kiểm tra
X_test_log = np.log(X_test)

# Xây dựng và huấn luyện mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train_log, y_train)

# Dự đoán giá trị 'MEDV' trên tập kiểm tra
y_pred = model.predict(X_test_log)

# Đánh giá mô hình
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
r2 = metrics.r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R^2:", r2)
```

```python
# Đơn biến
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt

# Load dataset
file_path = './boston.csv'
df = pd.read_csv(file_path)

# Select a single feature (LSTAT) and the target (MEDV)
X_single = df[['LSTAT']]
y = df['MEDV']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_single, y, test_size=0.2, random_state=1)

# Build and train the linear regression model
model_single = LinearRegression()
model_single.fit(X_train, y_train)

# Make predictions
y_pred_single = model_single.predict(X_test)

# Evaluate the model
rmse_single = np.sqrt(metrics.mean_squared_error(y_test, y_pred_single))
r2_single = metrics.r2_score(y_test, y_pred_single)

print("Đơn biến - RMSE:", rmse_single)
print("Đơn biến - R^2:", r2_single)

# Plotting actual vs predicted
plt.scatter(X_test, y_test, color="blue", label="Actual")
plt.plot(X_test, y_pred_single, color="red", label="Predicted")
plt.xlabel("LSTAT")
plt.ylabel("MEDV")
plt.title("Hồi quy tuyến tính đơn biến (LSTAT -> MEDV)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Tải dữ liệu
data = pd.read_csv('boston.csv')

# Xác định đặc trưng và mục tiêu
X = data[['AGE', 'LSTAT']]
y = data['MEDV']

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra theo tỷ lệ 80:20 với random state là 10
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Áp dụng biến đổi log cho các đặc trưng trong tập huấn luyện
X_train_log = np.log(X_train)

# Huấn luyện mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train_log, y_train)

# Áp dụng biến đổi log cho các đặc trưng trong tập kiểm tra
X_test_log = np.log(X_test)

# Dự báo giá trị MEDV trên tập kiểm tra
y_pred = model.predict(X_test_log)

# Tính toán lỗi bình phương trung bình
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Load the dataset
file_path = 'boston.csv'
df = pd.read_csv(file_path)

# Select features (AGE and LSTAT) and target (MEDV)
X = df[['AGE', 'LSTAT']]
y = df['MEDV']

# Split data into training and testing sets with an 80:20 ratio and random_state = 10
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Apply log transformation to the training data features
X_train_log = np.log(X_train)
X_test_log = np.log(X_test)

# Build and train the Linear Regression model
model = LinearRegression()
model.fit(X_train_log, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_log)

# Calculate R^2 on the test set
r2_test = metrics.r2_score(y_test, y_pred)

print("R^2 on the test set:", r2_test)
```

```python
import pandas as pd
from sklearn.preprocessing import MaxAbsScaler
```

```python

year = [2010,2011,2012,2013,2014,2015]
salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]
salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]

df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
```

```python
# Chuẩn hóa
# 1. Normalization
```

```python
# Tính max absolate scaler LA
max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la

df
```

```python
# Tính max absolute scaler HCM
max_hcm = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_hcm
df
```

```python
# Dùng 
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
#===PP1: Nếu tôi biết lương HCM và LA lần lượt là 4.5 và 5.75 thì qua scaler sẽ là bao nhiêu
# TH xuôi
vSc = scaler.transform([[4.5,5.75]])
vSc
```

```python
# Nếu tôi biết lương HCM và LA lần lượt là 0.5 và 0.8 thì qua scaler sẽ là bao nhiêu
# TH ngược
vR = scaler.inverse_transform([[0.5,0.8]])
vR
```

```python
#===PP2: min-max scale về đoạn [0, 1]

# tìm min, max của tập
# x - min / max -min 
```

```python
#===PP2: min-max scale về đoạn [0, 1]
min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```

```python
# HCM
min_hcm = df['HCM'].min()
max_hcm = df['HCM'].max()
df['mmcHCM'] = (df['HCM'] - min_hcm) / (max_hcm - min_hcm)

df
```

```python
#
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
# Lương HCM và LA lần lượt là 12 và 4.5
# Scaler
vSc = scaler.transform([[12,4.5]])
vSc
```

```python

# Trước khi biến đổi scaler thì gia trị bao nhiêu 0.8 và 1.8
# Inverse
vR = scaler.inverse_transform([[0.8,1.8]])
vR
```

```python
#=== 2. Standandization
# Ví dụ về Z-score
# Nếu 1 người có thu nhập 15 triệu thì ở LA hay TP HCM thoải mái hơn
# Nếu so sánh giá trị của 2 cột, không cùng hệ quy chiếu, 
```

```python
df[['HCM', 'LA']].describe()
```

```python
# Z-Score = x - muy / xích ma
# HCM
mean_hcm = df['HCM'].mean()
std_hcm = df['HCM'].std()
df['zHCM'] = (df['HCM'] - mean_hcm)
df['zHCM'] = df['zHCM'] / std_hcm
# df['zHCM']

# LA
mean_la = df['LA'].mean()
std_la = df['LA'].std()
df['zLA'] = (df['LA'] - mean_la)
df['zLA'] = df['zLA'] / std_la
df[['zHCM','zLA']]

```

```python
# Tính Zscore
mean_la = df['LA'].mean()
std_la = df['LA'].std()
df['zLA'] = (df['LA'] - mean_la) / std_la

df
```

```python
# -----------------------------
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)
print(scaled_df)
```

```python
# Một người 15tr ở HCM, 12 tr ở LA
# Zscore
vSc = scaler.transform([[15,12]])
vSc
```

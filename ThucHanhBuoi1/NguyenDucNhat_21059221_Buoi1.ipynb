```python
import pandas as pd
```

```python
#Colunm names
column_name = ['Id', 'Name', 'Age', 'Weight', 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']
df = pd.read_csv('patient_heart_rate.csv')
df.head(10)
```

```python
df[['Fisrtname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
df.head(10)
```

```python
# Chuyen don vi ve don ci chuan "kg"
weight = df['Weight']

for i in range(0, len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3:]:
       x = x[:-3:]
       float_x = float(x)
       y = int(float_x/2.2)
       #y = str(y)
       weight[i] = y
    if 'kgs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        weight[i] = x

df.rename(columns={'Weight': 'Weight(kgs)'}, inplace=True)  

df.head(10)
```

```python
# Xu ly dong co du lieu rong ( khong co du lieu co gia tri: NaN)
df.dropna(how='all', inplace=True)
df.head(10)
```

```python
# Xu ly dong co du lieu bi trung lap
df = df.drop_duplicates(subset=['Fisrtname', 'Lastname', 'Age', 'Weight(kgs)'])
df.head(10)
```

```python
# Doi ten cot Fisrtname thanh Firstname
df.rename(columns={'Fisrtname': 'Firstname'}, inplace=True)
```

```python
df.head(10)
```

```python
#Xu ly du lieu bi anh huong boi loi non-ASCII
df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(df)
```

```python
#9. Missing values
# Thay gia tri thieu cuar tuoi bang gia tri yeu vi
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
# print(df)
# Thay gia tri thieu cua can nang bang gia tri trung vi
df['Weight(kgs)'].fillna(df['Weight(kgs)'].median(), inplace=True)
# Ket qua sau khi xu ly cot Age va Weight
df.head(10)
```

```python
# 11. Phan ra du lieu

#Buoc 1: Tao melt du lieu de co cot gender_time
df = pd.melt(   
                df, id_vars =['Id', 'Age', 'Weight(kgs)','Firstname', 'Lastname'], 
                value_name= 'PulseRate', var_name='gender_time').sort_values(
                    [
                        'Id', 'Age', 'Weight(kgs)','Firstname', 'Lastname'
                    ]
                ) 
print(df)
             
```

```python
#Buoc 2: Tao dataframe tam thoi la ket qua cua viec tach cot gender_time
df_temp = df['gender_time'].str.extract('(\D)(\d+)(\d{2})', expand=True)
print(df_temp)
```

```python
#Buoc 3: Dat ten cho dataframe tam thoi
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
print(df_temp)
```

```python
#Buoc 4: Noi dataframe tam thoi voi dataframe ban dau
df = pd.concat([df, df_temp], axis=1)
print(df)
```

```python
#Buoc 5: Bo cot gender_time
df = df.drop(['gender_time'], axis=1)
df.head(5)
```

```python
# rename cot Weight(kgs) thanh Weight_kgs
df.rename(columns={'Weight(kgs)': 'Weight_kgs'}, inplace=True)
```

```python
# 12. Loai bo het cac dong du lieu thua ( la dong co phan PulseRate co dau -)
import numpy as np

df = df.replace('-', np.nan).dropna(subset=['PulseRate'])
df.head(10)
```

```python
# 13. Nhung ben nhan chua ghi nhan ho ten ( vi ly do nao do)
df['Firstname'].isnull().sum()
df['Lastname'].isnull().sum()
df.head(10)
```

```python
# Nhung gia tri nhip tim va thoi gian do nhip tim thi day du...
df['Firstname'].fillna('Unknown', inplace=True)
df['Lastname'].fillna('Unknown', inplace=True)
df.head(10)
```

```python
#14. reset index
df = df.reset_index()
df.head(10)
```

```python
#15. Luu tru du lieu vao file patient_heart_rate_clean.csv
df.to_csv('patient_heart_rate_clean.csv')
```

```python
# BAI TAP: PHAN TICH DU LIEU
# 1.  Vẽ biểu đồ Pie để thể hiện tỉ lệ bệnh nhân nam và nữ
# 2.  Hãy tạo cột Fullname = Firstname+’ ‘+Lastname
# 3.  Vẽ biểu đồ cột so sánh độ tuổi giữa các tên bệnh nhân
# 4.  Vẽ biểu đồ cột so sánh nhịp tim trung bình giữa các tên bệnh nhân
# 5.  Vẽ biểu đồ cột so sánh nhịp tim trung bình theo từng nhóm giờ 00-06, 06-12,12-18 có phân loại theo giới tính
# 6.  Vẽ biểu đồ phân tán thể hiện tương quan giữa trung bình độ tuổi và trung bình nhịp tim
# 7.  Vẽ biểu đồ đường thể hiện sự thay đổi của nhịp tim trung bình các bệnh nhân theo thứ tự khung giờ 00-06, 06-12,12-18 theo từng nhóm giới tính. 
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('patient_heart_rate_clean.csv')
# Loc data co id bi trung
data = data.groupby('Id').agg({
    'Age': 'mean',          
    'PulseRate': 'mean',    
    'Weight_kgs': 'mean',    
    'Gender': 'first',      
    'Firstname': 'first',    
    'Lastname': 'first',     
    'Lower_hour': lambda x: ','.join(map(str, x.unique())),  
    'Upper_hour': lambda x: ','.join(map(str, x.unique()))   
}).reset_index()

# Display the grouped data
data.head(13)
```

```python
# Cau 1: Vẽ biểu đồ Pie để thể hiện tỉ lệ bệnh nhân nam và nữ
# Count gender values
gender_counts = data['Gender'].value_counts()
# 
labels = gender_counts.index
sizes = gender_counts.values
colors = ['lightblue', 'lightcoral']  
explode = (0.1, 0)  
#
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Ty le benh nhan nam va nu')
plt.axis('equal') 
# show the plot
plt.show()

```

```python
# 2.  Hãy tạo cột Fullname = Firstname+’ ‘+Lastname
data['Fullname'] = data['Firstname'] + ' ' + data['Lastname']
data.head(13)
```

```python

# 3.  Vẽ biểu đồ cột so sánh độ tuổi giữa các tên bệnh nhân
data['Fullname'] = data['Firstname'] + ' ' + data['Lastname']

age_comparison = data.groupby('Lastname')['Age'].mean().sort_values()
plt.figure(figsize=(16, 10))
age_comparison.plot(kind='bar', color='skyblue')
plt.ylabel('Do tuoi trung binh')
plt.xlabel('Ten benh nhan')
plt.title('Do tuoi trung binh cua benh nhan theo ten')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show the plot
plt.tight_layout()  
plt.show()
```

```python
#4.  Vẽ biểu đồ cột so sánh nhịp tim trung bình giữa các tên bệnh nhân
pulse_rate_comparison = data.groupby('Fullname')['PulseRate'].mean().sort_values()
plt.figure(figsize=(16, 10))
pulse_rate_comparison.plot(kind='bar', color='skyblue')
plt.ylabel('Nhip tim trung binh')
plt.xlabel('Ten benh nhan')
plt.title('Nhip tim trung binh cua benh nhan theo ten')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 5.  Vẽ biểu đồ cột so sánh nhịp tim trung bình theo từng nhóm giờ 00-06, 06-12,12-18 có phân loại theo giới tính

grouped_data = data.groupby(['Lower_hour', 'Gender'])['PulseRate'].mean().unstack()
plt.figure(figsize=(16, 10))
grouped_data.plot(kind='bar', color=['lightblue', 'lightcoral'], width=0.8)

# Label the plot
plt.xlabel('Nhom gio')
plt.ylabel('Nhip tim trung binh')
plt.title('Nhip tim trung binh theo nhom gio va gioi tinh')
plt.xticks(rotation=0)  
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show the plot
plt.tight_layout()  
plt.show()


```

```python
# 6.  Vẽ biểu đồ phân tán thể hiện tương quan giữa trung bình độ tuổi và trung bình nhịp tim
# group by 
averages = data.groupby('Fullname').agg({'Age': 'mean', 'PulseRate': 'mean'})
# scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(averages['Age'], averages['PulseRate'], color='skyblue', edgecolor='black')
# Label the plot
plt.xlabel('Do tuoi trung binh')
plt.ylabel('Nhip tim trung binh')
plt.title('Bieu do phan tan giua do tuoi trung binh va nhip tim trung binh')
plt.grid(True)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 7. Vẽ biểu đồ đường thể hiện sự thay đổi của nhịp tim trung bình các bệnh nhân theo thứ tự khung giờ 00-06, 06-12,12-18 theo từng nhóm giới tính.
grouped_data = data.groupby(['Lower_hour', 'Gender'])['PulseRate'].mean().unstack()
plt.figure(figsize=(12, 8))
for gender in grouped_data.columns:
    plt.plot(grouped_data.index, grouped_data[gender], marker='o', label=gender)
plt.xlabel('Khung gio')
plt.ylabel('Nhip tim trung binh')
plt.title('Thay doi nhip tim trung binh theo khung gio va gioi tinh')
plt.legend()
plt.grid(True)
plt.show()
```
